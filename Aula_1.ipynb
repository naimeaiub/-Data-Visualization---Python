```python
import pandas as pd
```

```python
df = pd.read_csv('Dados/imigrantes_canada.csv', sep=',')
```

```python
df.head()
```

```python
df.info()
```

```python
df.set_index('País', inplace=True)
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
brasil = df.loc['Brasil', anos]
```

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Nº de imigrantes')
plt.xticks(['1980', '1985','1990','1995', '2000', '2005', '2010'])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))

ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
ax.set_title('Imigração do Brasil para o Canadá \n 1980 a 2013')
ax.set_xlabel('Ano')
ax.set_ylabel('Nº de imigrantes')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize=(16,4))

ax[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
ax[0].set_title('Imigração do Brasil para o Canadá \n 1980 a 2013')
ax[0].set_xlabel('Ano')
ax[0].set_ylabel('Nº de imigrantes')
ax[0].xaxis.set_major_locator(plt.MultipleLocator(5))
ax[0].grid()

ax[1].boxplot(dados_brasil['imigrantes'])
ax[1].set_title('Boxplot da imigração do Brasil para o Canadá \n 1980 a 2013')
ax[1].set_xlabel('Brasil')
ax[1].set_ylabel('Nº de imigrantes')
ax[1].grid()


plt.show()
```

```python
fig, ax = plt.subplots(2,2, figsize=(16,8))
fig.subplots_adjust(hspace=0.4, wspace=0.3)
fig.suptitle('Imigrações dos quatro paises para o Canadá \n 1980 a 2013')

ax[0,0].plot(df.loc['Brasil',anos])
ax[0,0].set_title('Imigração do Brasil para o Canadá \n 1980 a 2013')

ax[0,1].plot(df.loc['Colômbia',anos])
ax[0,1].set_title('Imigração da Colômbia para o Canadá \n 1980 a 2013')

ax[1,0].plot(df.loc['Argentina',anos])
ax[1,0].set_title('Imigração da Argentina para o Canadá \n 1980 a 2013')

ax[1,1].plot(df.loc['Peru',anos])
ax[1,1].set_title('Imigração da Peru para o Canadá \n 1980 a 2013')

for a in ax.flat:
    a.xaxis.set_major_locator(plt.MultipleLocator(5))

for a in ax.flat:
    a.set_xlabel('Ano')
    a.set_ylabel('Nº de imigrantes')

ymin = 0
ymax = 7000

for a in ax.ravel():
    a.set_ylim(ymin,ymax)

plt.show()
```
