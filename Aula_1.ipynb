```python
import pandas as pd
```

```python
df = pd.read_csv('Dados/imigrantes_canada.csv', sep=',')
```

```python
df.head()
```

```python
df.info()
```

```python
df.set_index('País', inplace=True)
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
brasil = df.loc['Brasil', anos]
```

```python
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
plt.title('Imigração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Nº de imigrantes')
plt.xticks(['1980', '1985','1990','1995', '2000', '2005', '2010'])
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize=(16,4))

ax[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
ax[0].set_title('Imigração do Brasil para o Canadá \n 1980 a 2013')
ax[0].set_xlabel('Ano')
ax[0].set_ylabel('Nº de imigrantes')
ax[0].xaxis.set_major_locator(plt.MultipleLocator(5))
ax[0].grid()

ax[1].boxplot(dados_brasil['imigrantes'])
ax[1].set_title('Boxplot da imigração do Brasil para o Canadá \n 1980 a 2013')
ax[1].set_xlabel('Brasil')
ax[1].set_ylabel('Nº de imigrantes')
ax[1].grid()


plt.show()
```

```python
dados_brasil.describe()
```

```python
fig, ax = plt.subplots(2,2, figsize=(16,8))
fig.subplots_adjust(hspace=0.4, wspace=0.3)
fig.suptitle('Imigrações dos quatro paises para o Canadá \n 1980 a 2013')

ax[0,0].plot(df.loc['Brasil',anos])
ax[0,0].set_title('Imigração do Brasil para o Canadá \n 1980 a 2013')

ax[0,1].plot(df.loc['Colômbia',anos])
ax[0,1].set_title('Imigração da Colômbia para o Canadá \n 1980 a 2013')

ax[1,0].plot(df.loc['Argentina',anos])
ax[1,0].set_title('Imigração da Argentina para o Canadá \n 1980 a 2013')

ax[1,1].plot(df.loc['Peru',anos])
ax[1,1].set_title('Imigração da Peru para o Canadá \n 1980 a 2013')

for a in ax.flat:
    a.xaxis.set_major_locator(plt.MultipleLocator(5))

for a in ax.flat:
    a.set_xlabel('Ano')
    a.set_ylabel('Nº de imigrantes')

ymin = 0
ymax = 7000

for a in ax.ravel():
    a.set_ylim(ymin,ymax)

plt.show()
```

```python
america_sul = df.query('Região == "América do Sul"')
```

```python
america_sul
```

```python
cores= ['royalblue','orange','forestgreen','orchid','purple','brown','slateblue',
        'gray','olive','navy','teal','tomato']

fig, ax = plt.subplots(figsize=(12,4))

ax.barh(america_sul.index,america_sul['Total'], color = cores)
ax.set_title('Imigração da América do Sul para o Canadá \n 1980 a 2013', loc = 'left')
ax.set_ylabel('')
ax.set_xlabel('Nº de imigrantes', fontsize = 14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show()
```

```python
america_sul_sorted = america_sul.sort_values(by='Total', ascending=True)
```

```python
cores= ['royalblue','orange','forestgreen','orchid','purple','brown','slateblue',
        'gray','olive','navy','teal','tomato']

fig, ax = plt.subplots(figsize=(12,4))

ax.barh(america_sul_sorted.index,america_sul_sorted['Total'], color = cores)
ax.set_title('Imigração da América do Sul para o Canadá \n 1980 a 2013', loc = 'left')
ax.set_ylabel('')
ax.set_xlabel('Nº de imigrantes', fontsize = 14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
plt.show()
```

```python
cores= []

for pais in america_sul_sorted.index:
    if 'Brasil' in pais:
        cores.append('green')
    else:
        cores.append('silver')

fig, ax = plt.subplots(figsize=(12,4))

ax.barh(america_sul_sorted.index,america_sul_sorted['Total'], color = cores)
ax.set_title('América do Sul: Brasil foi o quarto país com mais imigrantes \npara o Canadá no período de 1980 a 2013', loc = 'left')
ax.set_ylabel('')
ax.set_xlabel('Nº de imigrantes', fontsize = 14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

for i, v in enumerate(america_sul_sorted['Total']):
    ax.text(v + 20, i, str(v), color='black', fontsize=10, ha='left', va='center')

ax.set_frame_on(False)
# ax.get_xaxis().set_visible(False)
#ax.tick_params(axis='both',which='both',length=0)

fig.savefig('imigracao_america_sul_canada.png', transparent=False, dpi=300, bbox_inches='tight')
plt.show()
```
